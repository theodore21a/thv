```python
def knapsack(wt, val, W, n, t):
    # Base conditions
    if n == 0 or W == 0:
        return 0

    # Return the stored result if already computed
    if t[n][W] != -1:
        return t[n][W]

    # Choice diagram
    if wt[n - 1] <= W:
        t[n][W] = max(
            val[n - 1] + knapsack(wt, val, W - wt[n - 1], n - 1, t),
            knapsack(wt, val, W, n - 1, t)
        )
    else:
        t[n][W] = knapsack(wt, val, W, n - 1, t)

    return t[n][W]

def main():
    while True:
        print("\nMenu:")
        print("1. Enter item weights and profits")
        print("2. Enter knapsack capacity")
        print("3. Solve the knapsack problem")
        print("4. Exit")
        
        choice = input("Enter your choice: ")

        if choice == '1':
            global profits, weights
            profits = list(map(int, input("Enter profits of items (space-separated): ").split()))
            weights = list(map(int, input("Enter weights of items (space-separated): ").split()))
            if len(profits) != len(weights):
                print("Error: The number of profits and weights must be the same.")
                profits = weights = []
        elif choice == '2':
            global W
            W = int(input("Enter knapsack capacity: "))
        elif choice == '3':
            if not profits or not weights:
                print("Error: Enter item weights and profits first.")
                continue
            if 'W' not in globals():
                print("Error: Enter knapsack capacity first.")
                continue

            n = len(profits)
            t = [[-1 for _ in range(W + 1)] for _ in range(n + 1)]
            max_profit = knapsack(weights, profits, W, n, t)
            print(f"Maximum profit: {max_profit}")
        elif choice == '4':
            print("Exiting program.")
            break
        else:
            print("Invalid choice. Please try again.")

if __name__ == "__main__":
    profits = []
    weights = []
    W = 0
    main()
```
